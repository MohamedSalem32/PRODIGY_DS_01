## **1: Import Libraries**  
We begin by importing the required Python libraries for data analysis and visualization.  
- **Pandas**: For data loading and manipulation.  
- **Matplotlib**: For creating visualizations.  


```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

## **2: Load and Explore the Dataset**  
In this step, we will:  
1. Load the dataset from the CSV file.  
2. Explore the structure of the data by displaying the first few rows.  


```python
# Load the dataset
population_data = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_56.csv', skiprows=4)

# Display the first few rows
population_data.head()
```

## **3: Filter Data for Mauritania**  
We filter the dataset to focus solely on Mauritania's population data.  
- Extract rows where `Country Name` is "Mauritania".  
- Transpose the dataset to make years the index.  


```python
# Filter data for Mauritania
mauritania_data = population_data[population_data['Country Name'] == 'Mauritania']

# Keep only the columns from 1960 to 2023 and reset the index
mauritania_data = mauritania_data.loc[:, '1960':'2023']

# Transpose the data so that years are rows
mauritania_data = mauritania_data.T.reset_index()

# Rename columns appropriately
mauritania_data.columns = ['Year', 'Population']

# Ensure Year is of type integer and handle missing values
mauritania_data['Year'] = mauritania_data['Year'].astype(int)
mauritania_data = mauritania_data.dropna()

# Display the first few rows of the transformed DataFrame
mauritania_data.head()
```

## **4: Clean and Prepare the Data**  
In this step, we will:  
1. Handle missing values in the dataset.  
2. Ensure the `Year` column is of integer type for plotting.  


```python
# Handle missing values
mauritania_data = mauritania_data.dropna()

# Convert the Year column to integers
mauritania_data['Year'] = mauritania_data['Year'].astype(int)
mauritania_data.head()
```

## **5: Visualize the Population Distribution**  

We will create two visualizations to explore the population distribution of Mauritania:

1. **Bar Chart**: This will show the population distribution over time.
   - **X-axis**: Years
   - **Y-axis**: Population
   - **Chart Title**: "Population Distribution of Mauritania (1960-2023)"

2. We will create a line chart to visualize the population distribution of Mauritania over time.  
- X-axis: Years  
- Y-axis: Population  
- Chart Title: "Population Distribution of Mauritania (1960-2023)"  
"  


```python
# Plotting the population distribution using a bar chart
plt.figure(figsize=(12, 6))
plt.bar(mauritania_data['Year'], mauritania_data['Population'], color='b', label='Population')
plt.title('Population Distribution of Mauritania (1960-2023)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grid lines for y-axis
plt.legend()
plt.tight_layout()  # Adjust layout to avoid label clipping
plt.show()
```

```python
# Plotting the population distribution using a line chart 
plt.figure(figsize=(12, 6))
plt.plot(mauritania_data['Year'], mauritania_data['Population'], marker='o', color='b', label='Population')
plt.title('Population Distribution of Mauritania (1960-2023)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
